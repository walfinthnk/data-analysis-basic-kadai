```python
import numpy as np
import matplotlib.pyplot as plt
!pip install japanize-matplotlib
import japanize_matplotlib

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, tight_layout=True, figsize=(12, 8))

# Data for the first plot
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('折れ線グラフ')

# Data for the second plot (sine wave)
x_sin = np.linspace(0, 10, 100)
y1_sin = 2 + 2 * np.sin(2 * x_sin)
axes[0, 1].plot(x_sin, y1_sin, linewidth=2.0)
axes[0, 1].set_title('sin関数')
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
               ylim=(-1, 5), yticks=np.arange(-1, 6))

# Data for the third plot (histogram)
x_hist = np.random.normal(15, 5, 2000) # Use a new variable name for clarity
axes[0, 2].set_xlim(-10, 40)
axes[0, 2].hist(x_hist, bins=30)
axes[0, 2].set_title('ヒストグラム')

# Data for the fourth plot (scatter)
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# プロットする
axes[1,0].scatter(x,y)
axes[1,0].set_title('分布図')

# You can add more plots to other subplots if needed, for example:
# axes[1, 0].plot(...)
# axes[1, 1].plot(...)
# axes[1, 2].plot(...)

plt.show()
```
